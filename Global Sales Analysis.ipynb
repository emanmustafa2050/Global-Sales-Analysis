```python
import pandas as pd

# تحميل ملف Excel
file_path = "D:/Global Sales Analysis/Global Sales Analysis.xlsx"

# قراءة أسماء الشيتات في الملف
xls = pd.ExcelFile(file_path)
xls.sheet_names
```

```python
# قراءة البيانات من الشيت
df = pd.read_excel(xls, sheet_name="Sales Data")

# عرض أول 5 صفوف لفهم محتوى البيانات
df.head()
```

```python
# تنظيف البيانات

# 1. تحويل عمود التاريخ إلى نوع تاريخي صحيح
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"], errors="coerce")

# 2. إزالة الأعمدة غير المفيدة (ADDRESSLINE2 لأنها فارغة تقريبًا، و POSTALCODE لأنه ناقص)
df_cleaned = df.drop(columns=["POSTALCODE","Column1"])

# 3. التعامل مع القيم المفقودة في الأعمدة الأخرى
df_cleaned["STATE"] = df_cleaned["STATE"].fillna("Unknown")
df_cleaned["TERRITORY"] = df_cleaned["TERRITORY"].fillna("Unknown")


# عرض التغييرات بعد التنظيف
df_cleaned.head()
```

```python
df_cleaned.to_excel("Cleaned_Sales_Data.xlsx", index=False)
df_cleaned.to_csv("Cleaned_Sales_Data.csv", index=False)

```
